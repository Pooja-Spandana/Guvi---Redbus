```python
# importing mysql.connector 

import pandas as pd
import mysql.connector
```

```python
df = pd.read_csv(r'C:\Users\spand\Projects\REDBUS_APP\Redbus_data.csv')
df
```

```python
# SQL connection

client = mysql.connector.connect(
    host = "localhost",
    user ="root",
    password ="root",
    database = "MD_114"
    )

print("Connection successful!")
cursor = client.cursor()
```

```python
# SQL query to create table

cursor.execute('''CREATE TABLE IF NOT EXISTS Red_bus(
                  ID INT AUTO_INCREMENT PRIMARY KEY,
                  State_name TEXT,
                  Bus_name TEXT,
                  Bus_type TEXT,
                  Departure TIME,
                  Arrival TIME,
                  Duration TEXT,
                  Price DECIMAL(10, 2),
                  Ratings FLOAT,
                  Seats_Available INT,
                  Route_link TEXT,
                  Route_name TEXT  
                  )''')

print("Table Created successfully")
```

```python
# SQL query to insert data into Red_bus table

insert_query = '''INSERT INTO Red_bus(
                  State_name,
                  Bus_name,
                  Bus_type,
                  Departure,
                  Arrival,
                  Duration,
                  Price,
                  Ratings,
                  Seats_Available,
                  Route_link,
                  Route_name  
                  )
                  VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'''

data = df.values.tolist()

cursor.executemany(insert_query, data)

client.commit()

print("Values inserted successfully")
```
